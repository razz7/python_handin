```python
import csv

with open('data/netflix_titles.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        print(f'\t{row["show_id"]} type: {row["type"]} title is: {row["title"]} director: {row["director"]} .')
        line_count += 1
  #  print(f'Processed {line_count} lines.')
```
